```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("weight-height.csv")
df.head()
```

```python
sns.distplot(df["Height"])
```

```python
sns.boxplot(df["Height"]) #boxplot help us to find the outliers in our data
```

```python
df.shape
```

```python
df.describe()
```

```python
min_limit = df["Height"].quantile(0.01)
max_limit = df["Height"].quantile(0.99)
```

```python
max_limit
```

```python
min_limit
```

## Trimming

```python
new_data = df[(df["Height"]<max_limit) | (df["Height"]>min_limit)]
```

```python
new_data
```

```python
sns.distplot(new_data["Height"])
```

## Trimming

```python
new_data = df[(df["Height"]<max_limit) & (df["Height"]>min_limit)]
```

```python
sns.distplot(new_data["Height"])
```

```python
sns.boxplot(new_data["Height"])
```

## Capping

```python
new_data2 = df.copy() 
```

```python
new_data2["Height"] = np.where(df["Height"]>max_limit,max_limit, np.where(df["Height"]<min_limit,min_limit, df["Height"]))
```

```python
sns.distplot(new_data2["Height"])
```

```python
sns.boxplot(new_data2["Height"])
```
